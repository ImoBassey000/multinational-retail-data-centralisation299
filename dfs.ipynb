```python
import pandas as pd
from data_extraction import DataExtractor
from database_utils import DatabaseConnector
import tabula
import numpy as np
import re
```

```python

link = "https://data-handling-public.s3.eu-west-1.amazonaws.com/card_details.pdf"
dfs = tabula.read_pdf(link, pages='all')
dfs = pd.concat(dfs, ignore_index=True)
dfs.head()
```

```python
dfs.info()
```

```python
dfs['date_payment_confirmed'] = pd.to_datetime(dfs['date_payment_confirmed'], errors='coerce')
dfs = dfs.dropna(subset=['date_payment_confirmed'])
dfs['date_payment_confirmed'] = dfs['date_payment_confirmed'].dt.date
```

```python
var_1 = list(dfs['date_payment_confirmed'].unique())
var_1
```

```python
dfs = dfs.dropna(how='all')
dfs = dfs.drop_duplicates()
expiry_date_pattern = re.compile(r'^(0[1-9]|1[0-2])/([0-9]{2})$')
valid_expiry_dates = dfs['expiry_date'].apply(lambda x: bool(expiry_date_pattern.match(x)))
dfs = dfs[valid_expiry_dates]
```
